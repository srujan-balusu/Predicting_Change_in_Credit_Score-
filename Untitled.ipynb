# Predicting Change in Credit Score

## Part 1: Creating the Synthetic Dataset

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from imblearn.over_sampling import SMOTE
import shap
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1.1 Setting random seed and defining sample size

```python
np.random.seed(42)
n_samples=25000
```

### 1.2 Generating individual feature columns for every customer

```python
customer_id=[f"CUST_{i:05d}" for i in range(n_samples)]
age=np.random.randint(21,60,n_samples)
gender=np.random.choice(['Male','Female'],n_samples)
location=np.random.choice(['Urban','Semi-Urban','Rural'],n_samples)
monthly_income=np.random.randint(20000,300000,n_samples)
monthly_emi_outflow=(monthly_income*np.random.uniform(0.2,0.7,n_samples)).astype(int)
current_outstanding=np.random.randint(10000,500000,n_samples)
credit_limit=np.random.randint(50000,1000000,n_samples)
credit_utilization_ratio=np.minimum(current_outstanding/credit_limit,1.2)
num_open_loan=np.random.randint(1,8,n_samples)
repayment_history_score=np.random.randint(0,101,n_samples)
dpd_last_3_month=np.random.randint(0,91,n_samples)
num_hard_inquiry_last_6m=np.random.randint(0,6,n_samples)
recent_credit_card_usage=np.random.randint(1000,100000,n_samples)
recent_loan_disbursed_amount=np.random.randint(0,500000,n_samples)
total_credit_limit=np.random.randint(50000,1000000,n_samples)
months_since_last_default=np.random.randint(0,60,n_samples)
```

### 1.3 Defining target_credit_score_movement using heuristics

```python
target=[]
for i in range(n_samples):
    if (dpd_last_3_month[i]>60 and
        credit_utilization_ratio[i]>0.8 and
        num_hard_inquiry_last_6m[i]>2):
        target.append("decrease")
    elif (monthly_emi_outflow[i]/monthly_income[i]<0.3 and
          repayment_history_score[i]>=70):
        target.append("increase")
    else:
        target.append("stable")
```

### 1.4 Creating DataFrame

```python
df=pd.DataFrame({
    "customer_id":customer_id,
    "age":age,
    "gender":gender,
    "location":location,
    "monthly_income":monthly_income,
    "monthly_emi_outflow":monthly_emi_outflow,
    "current_outstanding":current_outstanding,
    "credit_limit":credit_limit,
    "credit_utilization_ratio":credit_utilization_ratio,
    "num_open_loan":num_open_loan,
    "repayment_history_score":repayment_history_score,
    "dpd_last_3_month":dpd_lat_3_month,
    "num_hard_inquiry_last_6m":num_hard_inquiry_last_6m,
    "recent_credit_card_usage":recent_credit_card_usage,
    "recent_loan_disbursed_amount":recent_loan_disbursed_amount,
    "total_credit_limit":total_credit_limit,
    "months_since_last_default":months_since_last_default,
    "target_credit_score_movement":target
})

df.head()
```

```python
print(df['target_credit_score_movement'].value_counts(normalize=True) * 100)
```

```python
n_decrease = 2000
dec_df = pd.DataFrame({
    "customer_id": [f"CUST_DEC_{i:05d}" for i in range(n_decrease)],
    "age": np.random.randint(21, 60, n_decrease),
    "gender": np.random.choice(['Male', 'Female'], n_decrease),
    "location": np.random.choice(['Urban', 'Semi-Urban', 'Rural'], n_decrease),
    "monthly_income": np.random.randint(20000, 300000, n_decrease),
    "monthly_emi_outflow": np.random.randint(10000, 100000, n_decrease),
    "current_outstanding": np.random.randint(300000, 500000, n_decrease),
    "credit_limit": np.random.randint(50000, 300000, n_decrease),
    "credit_utilization_ratio": np.random.uniform(0.9, 1.2, n_decrease),
    "num_open_loan": np.random.randint(4, 8, n_decrease),
    "repayment_history_score": np.random.randint(0, 60, n_decrease),
    "dpd_last_3_month": np.random.randint(61, 91, n_decrease),
    "num_hard_inquiry_last_6m": np.random.randint(3, 6, n_decrease),
    "recent_credit_card_usage": np.random.randint(1000, 100000, n_decrease),
    "recent_loan_disbursed_amount": np.random.randint(0, 500000, n_decrease),
    "total_credit_limit": np.random.randint(50000, 1000000, n_decrease),
    "months_since_last_default": np.random.randint(0, 60, n_decrease),
    "target_credit_score_movement": ["decrease"] * n_decrease
})
n_increase = 3000
inc_df = pd.DataFrame({
    "customer_id": [f"CUST_INC_{i:05d}" for i in range(n_increase)],
    "age": np.random.randint(21, 60, n_increase), 
    "gender": np.random.choice(['Male', 'Female'], n_increase),
    "location": np.random.choice(['Urban', 'Semi-Urban', 'Rural'], n_increase),
    "monthly_income": np.random.randint(100000, 300000, n_increase),
    "monthly_emi_outflow": np.random.randint(10000, 40000, n_increase),
    "current_outstanding": np.random.randint(10000, 150000, n_increase),
    "credit_limit": np.random.randint(300000, 1000000, n_increase),
    "credit_utilization_ratio": np.random.uniform(0.1, 0.4, n_increase),
    "num_open_loan": np.random.randint(1, 4, n_increase),
    "repayment_history_score": np.random.randint(80, 101, n_increase),
    "dpd_last_3_month": np.random.randint(0, 6, n_increase),
    "num_hard_inquiry_last_6m": np.random.randint(0, 2, n_increase),
    "recent_credit_card_usage": np.random.randint(1000, 100000, n_increase),
    "recent_loan_disbursed_amount": np.random.randint(0, 500000, n_increase),
    "total_credit_limit": np.random.randint(50000, 1000000, n_increase),
    "months_since_last_default": np.random.randint(0, 60, n_increase),
    "target_credit_score_movement": ["increase"] * n_increase
})

df_combined = pd.concat([df, dec_df, inc_df], ignore_index=True)
print(df_combined['target_credit_score_movement'].value_counts(normalize=True) * 100)
```

### 1.6 Saving to CSV

```python
df = df_combined
df.to_csv("synthetic_credit_score_dataset.csv", index=False)
```

## Part 2: Training & Evaluating the Model

### 2.1 Encoding Categorical Variables

```python
data=df.copy()
label_encoders={}
for col in['gender','location','target_credit_score_movement']:
    le=LabelEncoder()
    data[col]=le.fit_transform(data[col])
    label_encoders[col]=le

display(data[['gender','location','target_credit_score_movement']].head())
```

### 2.2 Splitting and Preprocessing the Data

```python
X=data.drop(columns=["customer_id","target_credit_score_movement"])
y=data["target_credit_score_movement"]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)

inv_map=label_encoders["target_credit_score_movement"].inverse_transform

print("Training Set Class Distribution:")
train_dist=y_train.value_counts(normalize=True).sort_index()
for i,val in train_dist.items():
    print(f"{inv_map([i])[0]}: {val:.2%}")

print("\nTest Set Class Distribution:")
test_dist=y_test.value_counts(normalize=True).sort_index()
for i,val in test_dist.items():
    print(f"{inv_map([i])[0]}: {val:.2%}")
```

### 2.3 Handling Class Imbalance using SMOTE

```python
sm=SMOTE(random_state=42)
X_train_res,y_train_res=sm.fit_resample(X_train,y_train)

print("Resampled Training Set Class Distribution:")
resampled_dist=y_train_res.value_counts(normalize=True).sort_index()
for i,val in resampled_dist.items():
    print(f"{label_encoders['target_credit_score_movement'].inverse_transform([i])[0]}: {val:.2%}")
```

### 2.4 Feature Scaling

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train_res)
X_test_scaled=scaler.transform(X_test)
```

### 2.5 Training the Random Forest Classifier

```python
rf=RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(X_train_scaled,y_train_res)
y_pred=rf.predict(X_test_scaled)

print(f"Number of features used: {X_train_scaled.shape[1]}")
```

### 2.6 Evaluating Model Performance

```python
print("Classification Report:")
target_names=label_encoders['target_credit_score_movement'].classes_
print(classification_report(y_test,y_pred,target_names=target_names,zero_division=0))
```

### 2.7 Visualizing Confusion Matrix

```python
cm=confusion_matrix(y_test,y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm,annot=True,fmt="d",cmap="Blues",
            xticklabels=target_names,yticklabels=target_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

### 2.8 Explaining Predictions Using SHAP

```python
explainer=shap.Explainer(rf,X_train_scaled)
shap_values=explainer(X_test_scaled[:200],check_additivity=False)
class_idx=2
shap_values_class=shap_values[...,class_idx]
shap.summary_plot(shap_values_class,features=X_test.iloc[:200],feature_names=X_test.columns)
shap.plots.bar(shap_values_class)
shap.plots.waterfall(shap_values_class[0])
```
